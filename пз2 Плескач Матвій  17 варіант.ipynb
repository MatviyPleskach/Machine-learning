6.Видаліть передостанній і останній стовпчики (Кількість кг хліба[20] на середню зарплату, Кількість БігМаків на середню зарплату).

7.Замінити тип нечислових даних на числові.

```python
data["Рік"] = data["Рік"].astype(str).str.replace("*", "")
data["Рік"] = data["Рік"].astype(int)
data["В національній валюті"] = data["В національній валюті"].str.replace('▲', "").str.replace('▼', "")
data["В доларах США"] = data["В доларах США"].str.replace('▲', "").str.replace('▼', "")
data['В національній валюті'] = data['В національній валюті'].str.replace(',', '.')
data['В доларах США'] = data['В доларах США'].str.replace(',', '.')
for col in data.iloc[:,1:]:
    if data[col].dtype != 'float64':
        print(data[col].dtype)
        data[col] = data [col].astype(float)
data
```

17.Обчисліть зарплату за останні два роки та порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку.

```python
last_2_years = data[data['Рік'].isin(data['Рік'].tail(2))]['В національній валюті']
last_5_years = data[data['Рік'].isin(data['Рік'].tail(5))]['В національній валюті']
sal_2 = data[data['Рік'].isin(last_2_years)]['В національній валюті']
sal_5 = data[data['Рік'].isin(last_2_years)]['В національній валюті']
if sal_2.mean() > sal_5.mean():
    print("Salaries have increased over the past two years") #incresed l2 than l5
elif sal_2.mean() < sal_5.mean():
    print("Salaries have decreased over the past two years") #decreased
else:
    print("Зарплата за останні два роки залишилася на тому ж рівні") #same lvl
```
