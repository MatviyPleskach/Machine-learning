Загрузити датасет, що містить інформацію про відвідуваність музеїв в різні роки.

Про які роки містить інформацію датасет.

```python
year = data['Date'].str.split('-').str[0].unique()
for year in year:
    print(year)
```

Змінити назви столбців. Замінити великі букви на маленьки, пробіли на підкреслення (date, avila_adobe, firehouse_museum, chinese_american_museum, america_tropical_interpretive_center).

```python
new_columns = {
    'Date' : 'date',
    'Avila Adobe' : 'avila_adobe',
    'Firehouse Museum' : 'firehouse_museum',
    'Chinese American Museum' : 'chinese_american_museum',
    'America Tropical Interpretive Center' : 'america_tropical_interpretive_center'
}
data.rename(columns=new_columns, inplace=True)
data
```

Обчислити середню кількість відвідувачів для кожного музею протягом всього періоду.

```python
vmean = round(data.drop(columns=['date']).mean())
print(vmean)
```

Знайти мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік

```python
df_2018 = data[data['date'].str.startswith('2018')]
min_visit = df_2018.drop(columns=['date']).min()
max_visit = df_2018.drop(columns=['date']).max()

print("Max visit", max_visit)
print("Min Visit",min_visit)
```

Порівняти кількість відвідувачів музею "Avila Adobe" у літні й зимові місяці 2018 року

```python
summer_m = [6,7,8]
winter_m= [12,1,2]
summer = df_2018[df_2018['date'].str.split('-').str[1].astype(int).isin(summer_m)]
winter = df_2018[df_2018['date'].str.split('-').str[1].astype(int).isin(winter_m)]

summer_visit = summer['avila_adobe'].sum()
winter_visit = winter['avila_adobe'].sum()

print("Visits on summer", summer_visit)
print("Visits on winter", winter_visit)
```

Знайти кореляцію між кількістю відвідувачів в кожному з музеїв та датами у 2016 році

```python
df_2016 = data[data['date'].str.startswith('2016')]
visit = ['avila_adobe','firehouse_museum','chinese_american_museum', 'america_tropical_interpretive_center']
visit_data = df_2016[visit]
corre = visit_data.corr()
print ("Correlation matrix visitors 2016:")
print(corre)
```

Побудувати теплокарту кореляції для попереднього завдання.

```python
plt.figure(figsize=(16,10))
sns.heatmap(corre ,annot = True, cmap='icefire')
plt.title("Correlation heatmap")
plt.show()
```

Побудувати графік ліній для кожного музею, показуючи їхню відвідуваність протягом 2017 року

```python
data_2017 = data[data['date'].str.startswith('2017')]
museum = ['avila_adobe','firehouse_museum','chinese_american_museum', 'america_tropical_interpretive_center']
visit = ['avila_adobe','firehouse_museum','chinese_american_museum', 'america_tropical_interpretive_center']
plt.figure(figsize=(12,6))
for museum in visit:
    sns.lineplot(data=data_2017, x='date', y=data_2017[museum], label=museum)
plt.title("@")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.legend(loc='upper right')
plt.show()
```

Побудувати графіки розсіювання для кожного музею за 2018 рік

```python
sns.pairplot(data=df_2018, vars=visit, height=2)
plt.suptitle("Pairplot Graphics")
plt.show()
```

Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий.

```python
df_2014 = data[data['date'].str.startswith('2014')]
df_2015 = data[data['date'].str.startswith('2015')]
fig,axes = plt.subplots(2,2, figsize =(12,6))
colors = ['blue','green','yellow','gray']
for i, data_year in enumerate([df_2014,df_2015,df_2016,data_2017]):
    summed_visit = data_year[visit].sum(axis=1)
    ax = axes[i//2, i % 2]
    sns.histplot(data=summed_visit,bins=30, kde=True, ax=ax, color=colors[i])
    ax.set_title(f"Year { i + 2014}")
    ax.set_xlabel("Summary visits")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
